### The needed imports 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import csv
%matplotlib inline
```

### Pruning the data to get rid of  invalid values 

```python
data = pd.read_csv('worldstat.csv', usecols=[0, 1, 2, 3, 4, 5, 6, 7, 8])
data = data.replace(to_replace='?.0', value='NaN')
data
```

### Plotting the distribution of values

```python
data.iloc[:,1].plot.hist(bins=100)
```

```python
data.iloc[:,2].plot.hist(bins=100)
```

### Calculating and printing the mean and standard deviation of the relevant columns 

```python
brm = data.describe().iloc[1,0]
brs = data.describe().iloc[2,0]
gdpm = data.describe().iloc[1,1]
gdps = data.describe().iloc[2,1]
lem = data.describe().iloc[1,2]
les = data.describe().iloc[2,2]
print("Birthrate mean: {:.4}, std: {:.4}".format(brm,brs))
print("GDP per capita mean: {:.7}, std: {:.7}".format(gdpm,gdps))
print("Life expetancy at birth mean: {:.5}, std: {:.3}".format(lem,les))
```

### Calculating the correlation between columns 

```python
data.iloc[:,[1,2]].corr(method='pearson')
```
